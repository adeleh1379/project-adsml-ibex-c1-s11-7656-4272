```python
import pandas as pd

# Load the dataset
file_path = 'p1_communes.csv'
df = pd.read_csv(file_path)
df
```

```python
# Display the number of rows and columns
num_rows, num_columns = df.shape
num_rows, num_columns
```

```python
# Display the first few entries of the DataFrame
first_entries = df.head()
first_entries
```

```python
# Obtain the index labels and column names
index_labels = df.index
column_names = df.columns.tolist()
index_labels, column_names
```

```python
# Check the data type for each column
data_types = df.dtypes
data_types
```

```python
# Check for missing values
missing_values = df.isnull().sum()
rows_with_missing_values = df[df.isnull().any(axis=1)]
missing_values
rows_with_missing_values
```

```python
# Remove observations with missing values
df_cleaned = df.dropna()
df_cleaned
```

```python
# Calculate the mean, minimum, and maximum value for each numerical column
numerical_columns = df_cleaned.select_dtypes(include=['int64', 'float64'])
mean_values = numerical_columns.mean()
min_values = numerical_columns.min()
max_values = numerical_columns.max()

mean_values.head(3)
```

```python
# List the 10 most populated communes
most_populated_communes = df_cleaned.nlargest(10, 'Residents')[['Commune', 'Residents']]
```

```python
# List the 10 least populated communes
least_populated_communes = df_cleaned.nsmallest(10, 'Residents')[['Commune', 'Residents']]
```

```python
# Group communes by canton and save them into separate CSV files
cantons = df_cleaned['Canton'].unique()
for canton in cantons:
    df_cleaned[df_cleaned['Canton'] == canton].to_csv(f'EPFL/{canton}.csv', index=False)

# List of created files
created_files = [f'{canton}.csv' for canton in cantons]
created_files
```

```python
# Compute the population density at the canton level and rank the cantons
canton_population_density = df_cleaned.groupby('Canton')['Residents'].sum() / df_cleaned.groupby('Canton')['Surface area in km²'].sum()
canton_population_density = canton_population_density.sort_values(ascending=False)
```

```python
# Convert the series to a DataFrame for easier display
canton_population_density_df = canton_population_density.reset_index(name='Population Density')
```

```python
canton_population_density_df 

# Observations:
# 1. Basel-Stadt: Has the highest population density, reflecting its status as a major urban center with a high concentration of residents in a small area.
# 2. Geneva: Also has a high population density, being an important international city with significant urban characteristics.
# 3. Zurich: As one of the largest and economically significant cities in Switzerland, Zurich has a high population density.
# 4. Other densely populated cantons include Zug and Basel-Landschaft, indicating significant urban development in these areas.
# 5. The rankings reflect the varying levels of urbanization and population distribution across the cantons, with major urban centers showing higher population densities.
```

```python
# Compute the number of communes in each canton where more than 50% of their population is aged between 20 and 64
communes_20_64 = df_cleaned[df_cleaned['20-64 years'] > 50].groupby('Canton').size().reset_index(name='Number of Communes')
```

```python
communes_20_64
```

```python
# Compute the difference between the maximum and minimum elevations for each canton
elevation_range = df_cleaned.groupby('Canton')['Elevation'].agg(lambda x: x.max() - x.min())
```

```python
# Find the top 5 cantons that have the largest range of elevations
top_5_elevation_range = elevation_range.nlargest(5).reset_index(name='Elevation Range')
```

```python
import matplotlib.pyplot as plt

# Get the top 10 populated communes using the previously calculated most populated communes for the plot

top_10_populated_communes = most_populated_communes.set_index('Commune')
```

```python
# Plotting the horizontal bar chart
plt.figure(figsize=(10, 8))
top_10_populated_communes['Residents'].plot(kind='barh', color='skyblue')
plt.title('Top 10 Populated Communes')
plt.xlabel('Population')
plt.ylabel('Commune')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest population at the top
plt.show()
```

```python
# Prepare the data for the horizontal stacked bar chart using df_cleaned
top_10_communes_cleaned = df_cleaned[df_cleaned['Commune'].isin(top_10_populated_communes.index)].copy()
```

```python
# Set the index to Commune
top_10_communes_cleaned.set_index('Commune', inplace=True)
area_types = ['Settlement area', 'Agricultural area', 'Wooded area', 'Unproductive area']
```

```python
# Normalize the area types to make sure they add up to 100%
top_10_communes_cleaned[area_types] = top_10_communes_cleaned[area_types].div(top_10_communes_cleaned[area_types].sum(axis=1), axis=0) * 100
```

```python
# Ensure to use only the top 10 communes
top_10_communes_cleaned = top_10_communes_cleaned.loc[top_10_populated_communes.index]
```

```python
# Plotting the horizontal stacked bar chart using the cleaned data
plt.figure(figsize=(12, 8))
top_10_communes_cleaned[area_types].plot(kind='barh', stacked=True, figsize=(12, 8), color=['blue', 'orange', 'green', 'red'])
plt.title('Land Use Distribution in Top 10 Populated Communes')
plt.xlabel('Percentage of Land Use')
plt.ylabel('Commune')
plt.legend(title='Land Use Types', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Observations:
# 1. Zurich: Predominantly composed of settlement area, indicating extensive urbanization with very little agricultural or unproductive land.
# 2. Geneva: High proportion of settlement area, similar to Zurich, reflecting its urban nature.
# 3. Basel: Mostly settlement area with minimal agricultural and wooded areas.
# 4. Lausanne: Significant settlement area, with some portions of agricultural and wooded land.
# 5. Bern: A more balanced land use with considerable settlement and agricultural areas, along with some wooded land.
# 6. Winterthur: Primarily settlement and agricultural land, with a notable wooded area.
# 7. Lucerne: High settlement area, with significant portions of agricultural and wooded land.
# 8. St. Gallen: Predominantly settlement and agricultural areas, reflecting urban and rural mix.
# 9. Lugano: Mostly settlement area, with notable agricultural and wooded land.
# 10. Biel/Bienne: High settlement area, with a mix of agricultural and wooded land.
```

```python
# Define the bins for the histogram
bins = [5, 10, 15, 20, 25, 30, 35]
labels = ['5%-10%', '10%-15%', '15%-20%', '20%-25%', '25%-30%', '30%-35%']
```

```python
# Create a copy of the cleaned dataframe
df_cleaned_copy = df_cleaned.copy()
```

```python

# Calculate the proportion of young residents (0-19 years) and create a new column
df_cleaned_copy['0-19 years proportion'] = df_cleaned_copy['0-19 years']

```

```python
# Bin the data based on the defined intervals
df_cleaned_copy['proportion_bin'] = pd.cut(df_cleaned_copy['0-19 years proportion'], bins=bins, labels=labels)
```

```python
# Plotting the histogram 
plt.figure(figsize=(5, 4))
df_cleaned_copy['proportion_bin'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Communes by Proportion of Young Residents (0-19 years)')
plt.xlabel('Proportion of Young Residents')
plt.ylabel('Number of Communes')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

# Group by languages
df_age_language = df_cleaned_copy[['0-19 years', 'Language']]

# Plotting the distributions using a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Language', y='0-19 years', data=df_age_language, palette='Set3')
plt.title('Distribution of Age Group 0-19 Years Across Language Regions')
plt.xlabel('Language Region')
plt.ylabel('Proportion of Age Group 0-19 Years')
plt.xticks(rotation=45)

# Adding legend
colors = sns.color_palette('Set3')
legend_labels = ['German-speaking', 'French-speaking', 'Romanian-speaking', 'Italian-speaking']
patches = [plt.plot([],[], marker="o", ms=10, ls="", mec=None, color=colors[i], 
            label="{:s}".format(legend_labels[i]))[0] for i in range(len(legend_labels))]
plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc='upper left', title="Language Regions")

plt.show()

# Observations:
# 1. German-speaking region: Shows a wide range of proportions for the age group 0-19 years, indicating diverse distributions of young residents.
# 2. French-speaking region: Narrower distribution, suggesting more uniformity in the proportion of young residents.
# 3. Italian-speaking region: Exhibits the widest range and a more spread out distribution, indicating significant variability in the proportion of young residents.
# 4. Romansh-speaking region: The distribution is quite compact, showing a consistent proportion of young residents across communes.
#These observations indicate that the proportion of young residents varies significantly across different language regions, with the German and Italian-speaking regions showing more variability compared to the French and Romansh-speaking regions.

```
